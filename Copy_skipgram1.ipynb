```python
# tokenizer = RegexpTokenizer(r'\w+')
guten_file_list = gutenberg.fileids()

def mapping(tokens):
    word_to_id = {}
    id_to_word = {}

    for i, token in enumerate(tokens):
        word_to_id[token] = i
        id_to_word[i] = token

    return word_to_id, id_to_word

def get_train_data(file_list, window, skipgram=True):

  vocab = []
  X = []
  Y = []

  for single_file in tqdm(file_list):
    sentences = gutenberg.sents(single_file)
    for sentence in sentences:
      clean_sentence = [word.lower() for word in sentence if word.isalpha() and word not in stop_words]
      vocab.extend(clean_sentence)
      x = []
      # y = []
      for idx in range(len(clean_sentence)-2*window):
        y = []
        sample = clean_sentence[idx : idx + 2*window+1]
        x = [sample[window]]*2*window
        for num in range(2*window+1):
          if num != window:
            y.append(sample[num])

        X.extend(x)
        Y.extend(y)

  #clean dataset
  #remove duplicate ordered pairs
  dataset = tuple(zip(X, Y))
  final_dataset = list(set(dataset))

  X = [data[0] for data in final_dataset]
  Y = [data[1] for data in final_dataset]

  #unique vocab words - need to maintain the order
  final_vocab = list(OrderedSet(vocab))

  word_to_id, id_to_word = mapping(final_vocab)

  if skipgram:
    return X, Y, word_to_id, id_to_word
  else:
    return Y, X, word_to_id, id_to_word
```

```python
window = 5

X, Y, word_to_id, id_to_word = get_train_data(guten_file_list, window)
```

```python
def one_hot_batch_encoding(batch, word_to_id):
  B = np.zeros((len(batch), len(word_to_id)))           #B_dim = bXV; batch_dim = [batch_size, 0]
  for i,w in enumerate(batch):
    B[i, word_to_id[w]] = 1
  return B
```

```python
class DataLoader:
  def __init__(self, word_to_id, id_to_word, X, Y, one_hot_encoding, batch_size):
    self.word_to_id = word_to_id
    self.id_to_word = id_to_word
    self.X = X
    self.Y = Y
    self.batch_size = batch_size
    self.batch_idx = 0
    self.max_length = len(X)
    self.one_hot_encoding = one_hot_encoding
  def __iter__(self):
    return self
  def __next__(self):
    curr_batch_idx = self.batch_idx
    self.batch_idx = self.batch_idx + self.batch_size
    if self.batch_idx < self.max_length:
      return self.one_hot_encoding(X[curr_batch_idx:self.batch_idx], self.word_to_id), self.one_hot_encoding(Y[curr_batch_idx:self.batch_idx], self.word_to_id)
    raise StopIteration
  def __len__(self):
    return len(self.X)//self.batch_size

```

```python
# loader = DataLoader(word_to_id, id_to_word, X, Y, one_hot_batch_encoding, batch_size = 1024)
loader = DataLoader(word_to_id, id_to_word, X, Y, one_hot_batch_encoding, batch_size = 100)
```

```python
def softmax(x):
  exp = np.exp(x - np.max(x))
  return exp/exp.sum()

def cross_entropy(y_pred, y_true, epsilon = 1e-9):
  return -np.mean(np.sum(y_true * np.log(y_pred + epsilon),axis=1))

def tss(y_pred, y_true):
  return 0.5*np.sum((y_pred - y_true)**2)

class word2vec:

  def __init__(self, vocab_size, embedding_size, output_activation, loss_fn):
    self.vocab_size = vocab_size
    self.embedding_size = embedding_size
    self.network = {
        "w1" : np.random.randn(self.vocab_size, self.embedding_size),
        "w2" : np.random.randn(self.embedding_size, self.vocab_size)
    }
    self.output_activation = output_activation
    self.loss_fn = loss_fn

  def forward(self, X):
    cache = {}
    cache["a1"] = X @ self.network["w1"]
    cache["a2"] = cache["a1"] @ self.network["w2"]
    cache["z"] = self.output_activation(cache["a2"])
    return cache

  def predict(self, X):
    return self.output_activation((X@self.network["w1"])@self.network["w2"])

  def backward(self, cache, x, y, alpha):
      # cache  = self.forward(x)
      da2 = cache["z"] - y
      dw2 = cache["a1"].T @ da2
      da1 = da2 @ self.network["w2"].T
      dw1 = x.T @ da1
      assert(dw2.shape == self.network["w2"].shape)
      assert(dw1.shape == self.network["w1"].shape)
      self.network["w1"] -= alpha * dw1
      self.network["w2"] -= alpha * dw2
      return self.loss_fn(cache["z"], y)
```

```python
model = word2vec(len(word_to_id), embedding_size = 100, output_activation = softmax, loss_fn = cross_entropy)
```

```python
L = []

epochs = 10

for epoch in range(epochs):
  print(f"Epoch : {epoch}")
  # loader = DataLoader(word_to_id, id_to_word, X, Y, one_hot_batch_encoding, batch_size = 500)
  with tqdm(total=len(loader)) as progress_bar:
    for x,y in loader:
      cache = model.forward(x)
      loss = model.backward(cache, x, y, alpha = 0.005)
      L.append(loss)
      progress_bar.update(1)
  with open(r'/content/drive/MyDrive/CS772/Assignment1/skipgram/win5_epochs10/SGram_model.pkl', 'wb') as f:
    dump(model, f)
  # print(f"Epoch : {epoch}")

with open("/content/drive/MyDrive/CS772/Assignment1/skipgram/win5_epochs10/loss.txt", "w") as f:
  for loss in L:
      f.write(str(loss))
```
